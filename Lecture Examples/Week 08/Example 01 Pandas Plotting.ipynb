```python
import pandas
```

```python
pandas.read_csv('BrisbaneWeather.csv')
```

```python
# Give the columns simpler names
pandas.read_csv('BrisbaneWeather.csv', names=['Month', 'MinTemp', 'MaxTemp', 'Rainfall'])
```

```python
# skip row 0 of the csv file as it is part of the header (containing column names)
pandas.read_csv('BrisbaneWeather.csv', names=['Month', 'MinTemp', 'MaxTemp', 'Rainfall'], header=0)
```

```python
# Use the Month column as our index
weather = pandas.read_csv('BrisbaneWeather.csv', names=['Month', 'MinTemp', 'MaxTemp', 'Rainfall'], header=0, index_col='Month')
weather
```

```python
# Demonstrating a simple plot (all numeric columns vs index)
weather.plot()
```

```python
# Plot just the MinTemp column 
weather.plot(y='MinTemp')
```

```python
# Plot a list of columns (vs index)
weather.plot(y=['MinTemp','MaxTemp'])
```

```python
# change the plot size to be 10 inches wide by 10 inches tall
weather.plot(y=['MinTemp','MaxTemp'], figsize=(10,10))
```

```python
# make the plot wider
weather.plot(y=['MinTemp','MaxTemp'], figsize=(20,8))
```

```python
# give the plot a title and a label for the y axis
weather.plot(y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title='Brisbane Weather')
```

```python
# add a grid to make it easier to correlate values to the x and y axis
weather.plot(y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title='Brisbane Weather', grid=True)
```

```python
# Specify that we want 12 labels on the x-axis (one for each month)
weather.plot(y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title='Brisbane Weather', grid=True, xticks=range(0,12))
```

```python
# Only show every other month label on the x-axis
weather.plot(y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title='Brisbane Weather', grid=True, xticks=range(0,12, 2))
```

```python
# Change the y-axis labels to go from 9 to 31 (degrees celcius)
weather.plot(y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title='Brisbane Weather', grid=True, xticks=range(0,12, 2), yticks=range(9,31))
```

```python
# Only show even numbered degrees on the y axis.
weather.plot(y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title='Brisbane Weather', grid=True, xticks=range(0,12, 2), yticks=range(8,31,2))
```

```python
# Use subplots to plot the two columns separately
weather.plot(y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title='Brisbane Weather', grid=True, xticks=range(0,12, 2), yticks=range(8,31,2), subplots=True)
```

```python
# Provide a list of titles, one for each subplot
weather.plot(y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title=['Brisbane Weather Mins', 'Brisbane Weather Maxs'], grid=True, xticks=range(0,12, 2), yticks=range(8,31,2), subplots=True)
```

```python
# back to only 1 subplot
weather.plot(y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title='Brisbane Weather', grid=True, xticks=range(0,12, 2), yticks=range(8,31,2))
```

```python
# By default the kind of plot is a 'line' plot
weather.plot(kind='line', y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title='Brisbane Weather', grid=True, xticks=range(0,12, 2), yticks=range(8,31,2))
```

```python
# Specifying kind='line' is equivalent to calling plot.line method
weather.plot.line(y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title='Brisbane Weather', grid=True, xticks=range(0,12, 2), yticks=range(8,31,2))
```

```python
# Plot a bar graph instead
weather.plot.bar(y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title='Brisbane Weather', grid=True, xticks=range(0,12, 2), yticks=range(8,31,2))
```

```python
# remove the grid and xticks
weather.plot.bar(y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title='Brisbane Weather', yticks=range(8,31,2))
```

```python
# Plot an 'area' graph instead
weather.plot.area(y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title='Brisbane Weather', yticks=range(8,31,2))
```

```python
# Set stacked to False so that the areas are not combined. Notice how the y axis now ranges from 0 to 30, not 0 to 50 as previously
weather.plot.area(y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title='Brisbane Weather', yticks=range(8,31,2), stacked=False)
```

```python
# The default is stacked=True which sums the 2 quantities producing a larger range - not really sensible for this example.
weather.plot.area(y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title='Brisbane Weather', yticks=range(0,51,2), stacked=True)
```

```python
# Create a box plot instead (see https://en.wikipedia.org/wiki/Box_plot for explanation)
weather.plot.box(y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title='Brisbane Weather', yticks=range(8,31,2), grid=True)
```

```python
# Create a histogram graph
weather.plot.hist(y=['MinTemp','MaxTemp'], figsize=(20,8), ylabel='Temperature (C)', title='Brisbane Weather', yticks=range(8,31,2), grid=True)
```

```python
# Remove the y axis label and ticks (since the y axis is now histogram frequency)
weather.plot.hist(y=['MinTemp','MaxTemp'], figsize=(20,8), title='Brisbane Weather', grid=True)
```

```python
# Group the x-axis values into 5 equal sized bins
weather.plot.hist(y=['MinTemp','MaxTemp'], figsize=(20,8), title='Brisbane Weather', grid=True, bins=5)
```

```python
# As previously, stacked=True results in the MinTemp and MaxTemp entries being added together
weather.plot.hist(y=['MinTemp','MaxTemp'], figsize=(20,8), title='Brisbane Weather', grid=True, bins=5, stacked=True)
```

```python
# Stacked=False results in the smaller values being displayed on top of the larger values, e.g. 2 min temps and 5 max temps in the range (17.5 to 21.5)
weather.plot.hist(y=['MinTemp','MaxTemp'], figsize=(20,8), title='Brisbane Weather', grid=True, bins=5, stacked=False)
```

```python
# Or use 15 smaller bins
weather.plot.hist(y=['MinTemp','MaxTemp'], figsize=(20,8), title='Brisbane Weather', grid=True, bins=15, stacked=False)
```

```python
# create a scatter plot of one variable vs another (a 12 x 12 'square' figure is better in this case)
weather.plot.scatter(x= 'MaxTemp', y='Rainfall', figsize=(12,12), title='Brisbane Weather', grid=True)
```

```python
# Finally, demonstrating how to create a pie chart (involes just one variable, Rainfall in this example)
weather.plot.pie(y='Rainfall', figsize=(12,12), title='Brisbane Weather')
```

```python
#  We can include the percentage in each slice by using the autopct parameter
weather.plot.pie(y='Rainfall', figsize=(12,12), title='Brisbane Weather', autopct='%.3f')
```

```python
# Show the percentage to 0 decimal places
weather.plot.pie(y='Rainfall', figsize=(12,12), title='Brisbane Weather', autopct='%.0f')
```

```python
# Add a percent '%' character to the end by adding '%%' to autopct 
weather.plot.pie(y='Rainfall', figsize=(12,12), title='Brisbane Weather', autopct='%.0f%%')
```

```python
# We don't need a legend in this case since all months are already labelled.
weather.plot.pie(y='Rainfall', figsize=(12,12), title='Brisbane Weather', autopct='%.0f%%', legend=False)
```

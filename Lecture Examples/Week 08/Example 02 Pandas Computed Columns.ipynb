```python
import pandas
```

```python
weather = pandas.read_csv('BrisbaneWeather.csv', names=['Month', 'MinTemp', 'MaxTemp', 'Rainfall'], header=0, index_col='Month')
weather
```

```python
# Access a specific column using indexing notation
weather['MinTemp']
```

```python
# Access the same column using the attribute notation
weather.MinTemp
```

```python
# Compute the difference between the MaxTemp and the MinTemp for each row. The result is a data series with the same (Month) index.
weather.MaxTemp - weather.MinTemp
```

```python
# Create a new column named 'Diff' that contains the difference between the MaxTemp and MinTemp for each row.
# Note that we can't use the attribute notation weather.Diff = ... when defining a new column that doesn't yet exist
weather['Diff'] = weather.MaxTemp - weather.MinTemp
```

```python
weather
```

```python
# Create a function that performs an ad hoc calculation based on a particular row from a Data Frame.
# This particular example uses only the name of the row (which is the Month name)
def days_in_month(row) :
    month = row.name
    if month == 'September' or month == 'April' or month == 'June' or month == 'November' :
        return 30
    elif month == 'February' :
        return 28
    else :
        return 31
```

```python
# Apply the days_in_month function to each row of the weather Data Frame to compute a new column named 'Days'
# Using axis=1 means apply the function to each row. If we instead used axis=0 then the function would be applied to each column of the Data Frame
weather['Days'] = weather.apply(days_in_month, axis=1)
weather
```

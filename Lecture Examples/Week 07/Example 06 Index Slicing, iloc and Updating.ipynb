```python
import pandas
pollution = pandas.read_csv('Example 01 Tabular Data/LSTM-Multivariate_pollution.csv', parse_dates=['date'], index_col='date')
```

```python
# retrieving a specific row and column based on row and column label
pollution.loc['2014-12-31 21:00:00', 'temp']
```

```python
# we can also use slicing to retrieve a range of columns
pollution.loc['2014-12-31 21:00:00', 'dew':'snow']
```

```python
# or using slicing to retrieve a range of rows and columns
pollution.loc['2014-12-31 00:00:00':'2014-12-31 23:59:00', 'dew':'snow']
```

```python
# all columns from the first, up to but not including snow
pollution.loc['2014-12-31 00:00:00':'2014-12-31 23:59:00', :'snow']
```

```python
# all columns from first to last
pollution.loc['2014-12-31 00:00:00':'2014-12-31 23:59:00', :]
```

```python
# or equivalently by omitting the column specification entirely
pollution.loc['2014-12-31 00:00:00':'2014-12-31 23:59:00']
```

```python
# all rows from the first, up to but not including '2014-12-31 23:59:00'
pollution.loc[:'2014-12-31 23:59:00', 'dew':'snow']
```

```python
# All rows from first to last. 
pollution.loc[:, 'dew':'snow']
```

```python
# In this case we can't omit the : as otherwise, dew and snow will be treated as row labels rather than column labels
pollution.loc['dew':'snow']
```

```python
# using step 5 to retrieve only every 5th row within that range
pollution.loc['2014-12-31 00:00:00':'2014-12-31 23:59:00': 5, 'dew':'snow']
```

```python
pollution
```

```python
# iloc is similar to loc, but we are using integers to index the rows and columns by row number and column number
# for example, retrieve row number 0 and column number 5 (which has name wnd_spd)
pollution.iloc[0, 5]
```

```python
# access the wind speed of the next row
pollution.iloc[1, 5]
```

```python
# use slicing to retrieve every 2nd row between rows 1 and 10
pollution.iloc[1:10:2, 2:]
```

```python
# retrieve every second row between first and last
pollution.iloc[::2]
```

```python
# we can also use iloc expressions on the right side of an assignment statement to update a DataFrame
pollution.iloc[0,5] = 1.9
```

```python
pollution
```

```python
# the update might apply to a range of rows (and/or columns)
pollution.iloc[0:10,5] = 1.9
```

```python
pollution
```

```python
#  we can also update using .loc expressions
pollution.loc["2014-12-31 21:00:00","temp"] = 1.9
```

```python
# let's try reading in a different data file ...
pollution = pandas.read_csv('Example 01 Tabular Data/LSTM-Multivariate_pollution.csv')
pollution
```

```python
# it works, but notice that the first column named "date" is of type object (which in this case really means string type)
pollution.info()
```

```python
# let's instead tell pandas to parse the strings in the date column as date time values
pollution = pandas.read_csv('Example 01 Tabular Data/LSTM-Multivariate_pollution.csv', parse_dates=['date'])
pollution
```

```python
#  the date column is now of type datetime64
pollution.info()
```

```python
pollution.date
```

```python
type(pollution.date)
```

```python
# when can then use the .dt attribute to access the constituent components of the date, for example the year of the date
pollution.date.dt.year
```

```python
# or the month number
pollution.date.dt.month
```

```python
# or the day number within the month
pollution.date.dt.day
```

```python
# get a complete list of all attributes and methods
dir(pollution.date.dt)
```

```python
# get a string that represents the name of the day
pollution.date.dt.day_name()
```

```python
# or the name of the month
pollution.date.dt.month_name()
```

```python
# we can also use these date components to filter, for example all months before June
pollution.date.dt.month < 6
```

```python
# pollution where the date is before June and the hour of the day is 11am (based on 24 hour clock)
pollution[(pollution.date.dt.month < 6) & (pollution.date.dt.hour == 11)]
```

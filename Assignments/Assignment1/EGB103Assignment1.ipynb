### Kilowatts of Refrigeration

The amount of refrigeration achieved by the chiller (measured in Kilowatts) is determined based on the amount by which the temperature of the water is reduced (measured in degrees Celsius) and the water flow rate (measured in litres per second). As described here (https://theengineeringmindset.com/chiller-cooling-capacity-calculate/), the Energy equation is:
$$Q = (\dot{V} \rho) c_p \Delta T$$
Q = Heat transfer kW
<br/>
$\dot{V}$ = Volume flow rate L/s
<br/>
$\rho$ = Density of Water
<br/>
$c_p$ = Specific Heat Capacity
<br/>
$\Delta$ T = temperature difference (Celcius or Kelvin)

In general the density and specific heat capacity depends on temperature, however we will assume a constant density of $1$ kg/litre and a specific heat capacity of $4.19$.

# Placing (x,y) values into bins

We can see from this scatter plot that there seems to some relationship between coefficient of performance and refrigeration load, however due to the large degree of natural variation, it is hard to see a clear pattern.

To see this performance curve more clearly we are going to divide the refrigeration loads into a sequence of finite ranges and then compute the average/mean coefficient of performance for each range. So, for example if each refrigeration range is of width 100kw,  then the first range (or bin) will be 0-100kw, the second range will be 100-200kw, the third from 200-300kw, etc.
Each of these ranges includes the lower bound but not the upper bound. So, for example, the range 100-200 contains all values $100 \le r < 200$. We refer to each of these ranges as a bin, as each (x,y) will be placed into one of these bins.

### Which bin?

Assuming that our minumum possible x value (refrigeration load) is zero, our first task is to be able to compute which bin to place a particular x value into ...

```python
# this function has been provided for you - do not modify

def calculate_bin_number(x_value, bin_size) :
    return int(x_value / bin_size)
```

### How many bins?

Next we need to determine how many bins we will need in total to span a particular list of x,y values, given a specified bin size ...

```python
# this function has been provided for  you - do not modify

def calculate_number_of_bins(xy_pairs, bin_size) :
    max_bin = -1
    for (x, y) in xy_pairs :
        bin = calculate_bin_number(x, bin_size)
        if bin > max_bin :
            max_bin = bin
    return max_bin + 1 # bins are numbers 0 .. max_bin, so the number of bins will be (max_bin+1)
```
